### Week 01, DNA sequence statistics (1)

First get the sequence with accession "MN908947" in the database "nuccore".

```r
getncbiseq <- function(db,accession) {
    require("rentrez")
    require("seqinr")
    require("XML")
    res_xml <- entrez_fetch(db=db, id=accession, rettype="fasta",retmode="XML")
    res_list <- xmlToList(res_xml)
    res_seq <- res_list[[1]][['TSeq_sequence']]
    res_vec <- tolower(s2c(res_seq))
    return(res_vec)
}
```

```r
ncovseq <- getncbiseq("nuccore","MN908947")
```
